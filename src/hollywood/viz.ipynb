```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from ast import literal_eval
```

```python
movies = pd.read_csv('movies_with_flag.csv')
movies['production_companies'] = movies['production_companies'].apply(lambda x: literal_eval(x))
```

```python
hof = movies[movies['is_hollywood_hof']]
```

```python
plt.hist(hof['year'], bins=20)
plt.suptitle("Nr. of 'hollywood' movies (HOF approach)")
```

```python
fig, axs = plt.subplots(2,2, figsize=(16,10))
fig.suptitle('share of missing data')

axs[0,0].set_title('all movies, revenue')
nan_counts = movies.groupby('year')['revenue'].apply(lambda x: (x.isna().sum())/len(x)).reset_index()
axs[0,0].bar(x=nan_counts['year'], height=nan_counts['revenue'])

axs[0,1].set_title('hollywood only, revenue')
nan_counts = hof.groupby('year')['revenue'].apply(lambda x: (x.isna().sum())/len(x)).reset_index()
axs[0,1].bar(x=nan_counts['year'], height=nan_counts['revenue'])

axs[1,0].set_title('all movies, budget')
nan_counts = movies.groupby('year')['budget'].apply(lambda x: (x.isna().sum())/len(x)).reset_index()
axs[1,0].bar(x=nan_counts['year'], height=nan_counts['budget'])

axs[1,1].set_title('hollywood only, budget')
nan_counts = hof.groupby('year')['budget'].apply(lambda x: (x.isna().sum())/len(x)).reset_index()
axs[1,1].bar(x=nan_counts['year'], height=nan_counts['budget'])
```
